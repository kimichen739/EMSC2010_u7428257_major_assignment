```python

# This code was generated by Gemini to turn a NetCDF file into a CSV file
# Open your uploaded Copernicus NetCDF file
# (Replace the name string if your file path differs)
ds = xr.open_dataset('barents_forecastalbedo.nc')

# Extract the forecast albedo variable and spatially average it
# This takes all grid pixels inside your Barents Sea box and averages them per month
# In ERA5, the forecast albedo variable is shorthand labeled as 'fal'
spatial_mean = ds['fal'].mean(dim=['latitude', 'longitude'])

# Convert the clean multi-decade timeline directly to a standard Pandas DataFrame
df_albedo = spatial_mean.to_dataframe(name='albedo').reset_index()

# Clean up columns: Keep only the time dimension and your calculated albedo
# Note: ERA5 time variables are often named 'valid_time' or 'time'
df_albedo = df_albedo[['valid_time', 'albedo']]
df_albedo.rename(columns={'valid_time': 'date'}, inplace=True)

# Export to a flat CSV file right into your project directory
df_albedo.to_csv('barents_sea_albedo.csv', index=False)

print("Success! NetCDF successfully converted to 'barents_sea_albedo.csv'")
print(df_albedo.head)
```
